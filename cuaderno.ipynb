```python
import requests
import pandas as pd

# 1. Configuración de la API
API_KEY = "DNvhatrOq3bfsukr3MKJOgBA1NjyMCNlUHvW9oya"  # Api_key de https://developer.nrel.gov/docs/"
URL = "https://developer.nrel.gov/api/alt-fuel-stations/v1.json"

# 2. Parámetros de la consulta
# Queremos: Solo estaciones eléctricas (ELEC), en el estado de Washington (WA)
params = {
    'api_key': API_KEY,
    'fuel_type': 'ELEC',
    'state': 'WA',
    'status': 'E',  # Solo estaciones operativas (E = Existing)
    'limit': 'all'  # Queremos todos los registros disponibles
}

# 1. Petición de datos
response = requests.get(URL, params=params)
data = response.json()

# 2. Convertir la lista de estaciones a un DataFrame
df_api = pd.DataFrame(data['fuel_stations'])


# Seleccionamos solo las que suelen ser fijas y útiles
columnas_utiles = ['station_name', 'city', 'zip', 'latitude', 'longitude', 'ev_connector_types']
df_api_limpio = df_api[columnas_utiles].copy()

# 4. Ver el resultado
print(f"Filas descargadas: {len(df_api_limpio)}")
print(df_api_limpio.head())

# Guardar para el proyecto
df_api_limpio.to_csv("datos_api_nrel.csv", index=False)
```
